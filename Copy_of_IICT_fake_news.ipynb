```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("fakenews_mastersheet.csv")
```

```python
print(df.head())
print(df.columns)
print(df.shape)
print(df.info())
```

```python
df.isnull().sum()
```

```python
df["label"].value_counts()
```

```python
df.duplicated(subset=["title","text"]).sum()
```

```python
df = df.drop_duplicates(subset=["title", "text"])
```

```python
df["title"] = df["title"].fillna("")
df["text"] = df["text"].fillna("")
```

```python
df["content"] = df["title"] + " " + df["text"]

X = df["content"]
y = df["label"]
```

```python
df["clean_content"] = df["content"]
```

```python
print(df.columns.tolist())
```

```python
df["clean_content"] = df["clean_content"].str.lower()
```

```python
df["clean_content"] = df["clean_content"].str.replace(r"http\S+|www\S+", "", regex=True)
```

```python
df["clean_content"] = df["clean_content"].str.replace(r"<.*?>", "", regex=True)
```

```python
df["clean_content"] = df["clean_content"].str.replace(r"[^\w\s]", "", regex=True)
```

```python
df["clean_content"] = df["clean_content"].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

df["clean_content"] = df["clean_content"].apply(
    lambda text: " ".join(
        word for word in text.split() if word not in stop_words
    )
)
```

```python
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize

df["tokens"] = df["clean_content"].apply(word_tokenize)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
